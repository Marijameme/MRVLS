```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import nbformat 
from importlib import import_module
```

```python
num_columns = [5,7,10,12,15]
bf_times = np.load('bf_times.npy')
ga_times = np.load('ga_times_small.npy')
sa_times = np.load('sa_times_small.npy')
```

```python
plt.plot(num_columns, bf_times, marker='o', color='#FFB400', label='brute force')
plt.plot(num_columns, ga_times, marker='o', color='#FF7C15', label='genetic algorithm')
plt.plot(num_columns, sa_times, marker='o', color='#ED4672', label='simulated annealing')

plt.legend(loc='center right',fancybox=True, ncol=1)
plt.grid(True)

plt.xlabel('Number of columns')
plt.ylabel('Time to compute')
plt.title('Time to compute vs number of columns')
```

```python
ga_big = np.load('ga_times_big.npy')
sa_big = np.load('sa_times_big.npy')
```

```python
plt.plot(num_columns, ga_big, marker='o', color='#FF7C15', label='genetic algorithm')
plt.plot(num_columns, sa_big, marker='o', color='#ED4672', label='simulated annealing')

plt.legend(loc='upper left',fancybox=True, ncol=1)
plt.grid(True)

plt.xlabel('Number of columns')
plt.ylabel('Time to compute')
plt.title('Time to compute vs number of columns')
```
